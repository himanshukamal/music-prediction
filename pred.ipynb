```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data_Train.csv',encoding='utf-8')
df
```

```python
df.drop('Name' ,axis = 1, inplace = True)
df.drop('Country',axis=1, inplace=True)
df.drop('Song_Name' ,axis = 1, inplace = True)
```

```python
df
```

```python
df.drop('Unique_ID' ,axis = 1, inplace = True)
```

```python
dx=df[:40000]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
list1=['Genre']
for val in list1:
    dx[val]=le.fit_transform(dx[val].astype(str))
```

```python
dx
```

```python
dx.drop('Timestamp' ,axis = 1, inplace = True)
```

```python
dx
```

```python
dx['Popularity']=dx['Popularity'].str.replace(',','')
dx['Popularity'] = (dx['Popularity'].replace(r'[KM]+$', '',regex=True).astype(float) * \
              dx['Popularity'].str.extract(r'[\d\.]+([KM]+)', expand=False)
                .fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
dx['Likes']=dx['Likes'].str.replace(',','')
dx['Likes'] = (dx['Likes'].replace(r'[KM]+$', '',regex=True).astype(float) * \
              dx['Likes'].str.extract(r'[\d\.]+([KM]+)', expand=False)
                .fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int))
```

```python
dx
```
